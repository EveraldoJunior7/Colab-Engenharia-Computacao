```python
def f(t, y):
  return y * np.sin(t) ** 3
```

```python
def edo(x, c=np.exp(2) ** (1/3)):
   return c * np.exp(-np.cos(x) + np.cos(x) ** 3 / 3)
```

```python
def y_real(a, b, h):
  return list(edo(np.arange(a, b + h, h)))
```

```python
def euler(t, y, a, b, h):
  ti = [t]
  yi = [y]
  real = [y]
  e = [0]
  while(ti[-1] < b):
    yi += [yi[-1] + h * f(ti[-1], yi[-1])]
    ti += [ti[-1] + h]

    real += [edo(ti[-1])]
    e += [abs(((yi[-1] - real[-1]) / real[-1]) * 100)]

  return yi
  #return pd.DataFrame(data={'ti': ti, 'yi': yi, 'y': real, 'erro (%)': e})
```

```python
euler(0.0, 1.0, 0.0, 3.0, 0.1)
```

```python
def ponto_medio(t, y, a, b, h):
  ti = [t]
  yi = [y]
  real = [y]
  e = [0]
  y_meio, t_meio = ['-'], ['-']
  while(ti[-1] < b):
    y_meio += [yi[-1] + f(ti[-1], yi[-1]) * (h / 2)]
    t_meio += [(ti[-1] + ti[-1] + h) / 2]
    yi += [yi[-1] + f(t_meio[-1], y_meio[-1]) * h]
    ti += [ti[-1] + h]
    real += [edo(ti[-1])]
    e += [abs(((yi[-1] - real[-1]) / real[-1]) * 100)]

  return yi
  #return pd.DataFrame(data={'ti': ti, 'yi+1/2': y_meio, 'yi': yi, 'y': real, 'erro (%)': e})
```

```python
ponto_medio(0.0, 1.0, 0.0, 3.0, 0.1)
```

```python
def runge_kutta_4(t, y, a, b, h):
  ti = [t]
  yi = [y]
  real = [y]
  e = [0]
  k1, k2, k3, k4 = ['-'], ['-'], ['-'], ['-']
  while(ti[-1] < b):
    k1 += [f(ti[-1], yi[-1])]
    k2 += [f(ti[-1] + (h / 2), yi[-1] + 0.5 * k1[-1] * h)]
    k3 += [f(ti[-1] + (h / 2), yi[-1] + 0.5 * k2[-1] * h)]
    k4 += [f(ti[-1] + h, yi[-1] + k3[-1] * h)]

    yi += [yi[-1] + (1/6) * (k1[-1] + 2 * k2[-1] + 2 * k3[-1] + k4[-1]) * h]
    ti += [ti[-1] + h]

    real += [edo(ti[-1])]
    e += [abs(((yi[-1] - real[-1]) / real[-1]) * 100)]

  return yi
  #return pd.DataFrame(data={'ti': ti, 'k1': k1, 'k2': k2, 'k3': k3, 'k4': k4, 'yi': yi, 'y': real, 'erro (%)': e})
```

```python
runge_kutta_4(0.0, 1.0, 0.0, 3.0, 0.1)
```

```python
x, y, a, b, h = 0.0, 1.0, 0.0, 3.0, .1

xi = np.arange(a, b + h, h)

fig, ax = plt.subplots()
fig.set_dpi(100)
plt.plot(np.arange(a, b + 0.1, 0.1), edo(np.arange(a, b + 0.1, 0.1)), linewidth=3, c='r')
plt.plot(xi, euler(x, y, a, b, h), '-x', linewidth=.5)
plt.plot(xi, ponto_medio(x, y, a, b, h), '-x')
plt.plot(xi, runge_kutta_4(x, y, a, b, h), '-x', linewidth=.75)
plt.xticks(np.arange(0, 3.25, step=0.25), rotation=90)
plt.grid(axis='both', alpha=.3)
plt.gca().spines["top"].set_alpha(0.0)
plt.gca().spines["right"].set_alpha(0.0)
plt.legend(['Verdadeiro', 'Euler', 'Ponto M√©dio', 'Runge-Kutta $(4^a ordem)$'])
fig.suptitle(f'Compara√ß√£o dos m√©todos, h={h}')
ax.set(xlabel='t', ylabel='y', title=f"$y'=y sen^3(t), y(0)=1, t‚àä[0,3]$")
plt.show()
```
