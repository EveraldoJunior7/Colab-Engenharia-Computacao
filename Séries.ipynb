```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from google.colab import widgets, output
from six.moves import zip
```

```python
def f(x):
  return sp.log(x)
```

```python
def df(xi, n):
  x = sp.Symbol('x')
  dfdx = sp.diff(f(x), x, n)
  return sp.lambdify([x], dfdx)(xi)
```

```python
def taylor(x, c, n):
  somatorio = 0
  for k in range(0, n + 1):
    somatorio += df(c, k) * (x - c) ** k / np.math.factorial(k)
  return somatorio
```

```python
print(taylor(0.5, 1, 100))
print(np.log(0.5))
```

# Euler {$e^{x}$}

```python
def euler(x, i):
  soma = 0
  for n in range(0, i + 1):
    soma += (x ** n) / np.math.factorial(n)
  return soma
```

```python
def criar_tab_euler(a, b):
  tabs = [str(i) for i in range(a,b)]
  tb = widgets.TabBar(tabs, location='top')
  x = np.arange(-3, 3, 0.1)
  for i in range(a, b):
    with tb.output_to(str(i)):
      fig, ax = plt.subplots()
      fig.set_dpi(100)
      plt.plot(x, np.exp(x), c='red', linewidth=2.0)
      plt.plot(x, euler(x, i), c='blue', linestyle='-.')
      plt.grid(axis='both', alpha=.3)
      plt.gca().spines["top"].set_alpha(0.0)
      plt.gca().spines["right"].set_alpha(0.0)
      plt.legend(['e^x', f'Taylor ordem {i}'])
      ax.set(xlabel='x', ylabel='y', ylim=(-2.5, 17.5), title=f'Comparação dos Gráficos')

criar_tab_euler(0, 10)
```

# Logaritimo Natural {$ln(x)$}

```python
def ln(x, i):
  soma = 0
  for n in range(1, i + 1):
    soma += (-1) ** (n - 1) * ((x - 1) ** n) / n
  return soma
```
